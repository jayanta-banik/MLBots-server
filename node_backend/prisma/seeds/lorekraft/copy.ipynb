```python
import pyperclip
from tqdm.auto import tqdm
```

```python
with open("data.txt", "r") as file:
    seeds = file.read()
```

```python
seeds = seeds.split("\n\n\n\n")
```

```python
len(seeds)
```

```python
for i in tqdm(seeds):
    input("Copied to clipboard. Press Enter to copy the next one...")
    print("copied: ", i[:50])
    pyperclip.copy(i)
```
