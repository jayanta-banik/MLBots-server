```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd
```

```python
URL = "https://www.topuniversities.com/world-university-rankings?items_per_page=200&tab=indicators&region=North%20America&countries=us&sort_by=rank&order_by=asc"


options = Options()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(options=options)
```

```python
driver.get(URL)
```

```python
container = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.ID, "ranking-data-load")))
container.text
```

```python
container = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.ID, "ranking-data-load")))

indicator_rows = [
    i.text.strip()
    for i in container.find_elements(
        By.XPATH,
        ".//div[@class='row']",
    )
    if i.text.strip() and len(i.text) > 5
]
```

```python
mylist = pd.DataFrame([i.split("\n") for i in indicator_rows])
```

```python
mylist = mylist.iloc[:, :3]
```

```python
mylist.columns = ["Rank", "University", "Location"]
```

```python
mylist.drop_duplicates(inplace=True)
```

```python
def is_non_numeric(x):
    try:
        float(x)
        return False
    except (ValueError, TypeError):
        return True


mylist = mylist[mylist["University"].apply(is_non_numeric)]
```

```python
mylist = mylist[mylist.University.apply(lambda x: not x.isdigit())]
```
