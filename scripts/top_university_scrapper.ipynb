```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
URL = "https://www.topuniversities.com/world-university-rankings?items_per_page=150&tab=indicators&region=North%20America&countries=us&sort_by=rank&order_by=asc"


options = Options()
options.binary_location = "/usr/bin/chromium"
options.add_argument("--headless=new")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(options=options)
```

```python
driver.get(URL)

container = WebDriverWait(driver, 30).until(
    EC.presence_of_element_located((By.ID, "newRankingFilters_newUI"))
)

indicator_rows = container.find_elements(
    By.CSS_SELECTOR,
    "div.row.ind-row.hide-this--in-mobile-indi",
)

print(f"Found {len(indicator_rows)} matching rows inside #newRankingFilters_newUI")
for index, row in enumerate(indicator_rows, start=1):
    text = row.text.strip().replace("\n", " ")
    print(f"{index}. {text[:200]}")

indicator_rows
```
